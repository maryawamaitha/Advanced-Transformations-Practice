# Clean the files and combine them into one final DataFrame.

```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
df=pd.read_csv(r"C:\Users\marya\Downloads\superhero_info - superhero_info.csv")
df.head()
```

```python
## adding expand=True
df['Hero|Publisher'].head(2)
```

#  Splitting the Hero and Publisher columns

```python
## save the 2 new columns into the dataframe
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
df.head(2)
```

```python
## drop the original column 
df = df.drop(columns=['Hero|Publisher'])
df.head(2)
```

```python
## examining a single value from the coordinates col
heigh = df.loc[0,"Measurements"]
print(type(heigh))
heigh
```

```python
## use .str.replace to replace all single quotes
df['Measurements'] = df['Measurements'].str.replace("'",'"').str.replace(" cm","").str.replace(" kg","")
## Apply the json.loads to the full column
df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements'].head()
```

```python
heigh_weigh = df['Measurements'].apply(pd.Series)
heigh_weigh
```

```python
df = pd.concat((df, heigh_weigh), axis = 1)
df.head(10)
```

```python
df = df.drop(columns=['Measurements'])
df.head(2)
```

```python
df2=pd.read_csv(r"C:\Users\marya\Downloads\superhero_powers - superhero_powers.csv")
df2.head()
```

```python
df2.drop(df2[df2['hero_names'] == '3-D Man'].index, inplace=True)
```

```python
df2.loc[1,'Powers']
```

```python
df2.sort_values(['hero_names'])
```

```python
df2["Powers"] = df2["Powers"].str.split(',')
```

```python
## exploding the column of lists
exploded = df2.explode("Powers")
exploded
```

```python
cols_to_make = exploded.dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df2[col] = df2['Powers'].str.contains(col)
df2.head()

```

```python
df2 = df2.drop(columns=['Powers'])
df2.head(2)
```

```python
df3 = pd.concat((df,df2), axis = 1)
df3.head(10)
```
